```python
page = requests.get(link)
```

```python
soup = bs(page.content, 'html.parser')
#it gives us the visual representation of data
print(soup.prettify())
```

```python
name=soup.find('span',class_="B_NuCI")
print(name)
```

```python
rating=soup.find('div',class_="_3LWZlK")
print(rating)
rating.text
```

```python
#get other details and specifications of the product
specification=soup.find('div',class_="_2cM9lP")
print(specification)
specification.text
```

```python
#get price of the product
price=soup.find('div',class_='_30jeq3 _16Jk6d')
print(price)
price.text
```

```python
products=[]              #List to store the name of the product
prices=[]                #List to store price of the product
ratings=[]               #List to store rating of the product
apps = []                #List to store supported apps                
os = []                  #List to store operating system
hd = []                  #List to store resolution
sound = []    
```

```python
for data in soup.findAll('div',class_={"_1YokD2 _3Mn1Gg col-8-12"}):
        names=data.find('div', attrs={'class':'B_NuCI'})
        price=data.find('div', attrs={'class':'_30jeq3 _16Jk6d'})
        rating=data.find('div', attrs={'class':'_3LWZlK'})
        specification = data.find('div', attrs={'class':'_2cM9lP'})
        
        for each in specification:
            col=each.find_all('div', class_="_1AtVbE col-12-12")
            app =data.find('li',class_="_21lJbe")
            os_ = data.find('li',class_="_21lJbe")
            hd_ = data.find('li',class_="_21lJbe")
            sound_ = data.find('li',class_="_21lJbe")
        products.append(names) # Add product name to list
        prices.append(price) # Add price to list
        apps.append(app)# Add supported apps specifications to list
        os.append(os) # Add operating system specifications to list
        hd.append(hd) # Add resolution specifications to list
        sound.append(sound) # Add sound specifications to list
        ratings.append(rating)   #Add rating specifications to list
```

```python
#printing the length of list
print(len(products))
print(len(ratings))
print(len(prices))
print(len(apps))
print(len(sound))
print(len(os))
print(len(hd))
```

```python
import pandas as pd
df=pd.DataFrame({'Product Name':products,'Supported_apps':apps,'sound_system':sound,'OS':os,"Resolution":hd,'Price':prices,'Rating':ratings})
df.head(10)
```
